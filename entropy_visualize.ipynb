# Trực quan hóa hàm cross entropy


```python
import numpy as np
import matplotlib.pyplot as plt

# Tính cross entropy theo a và y
def _cross_entropy(a, y):
    return -(y * np.log(a) + (1 - y) * np.log((1 - a)))

# Tính bình phương khoảng cách
def _squared_distance(a, y):
    return (a - y) ** 2

# Khởi tạo gía trị a từ 0 đến 1
a = np.linspace(0.001, 0.999, 200)

# Hàm visualize cross entropy và bình phương khoảng cách
def _plot_crs(a, y1, y2):
    cross_entropy1 = _cross_entropy(a, y1)
    squared_distance1 = _squared_distance(a, y1)
    
    min_entropy_index1 = np.argmin(cross_entropy1)
    min_entropy_value1 = cross_entropy1[min_entropy_index1]
    min_entropy_a1 = a[min_entropy_index1]
    
    min_distance_index1 = np.argmin(squared_distance1)
    min_distance_value1 = squared_distance1[min_distance_index1]
    min_distance_a1 = a[min_distance_index1]

    cross_entropy2 = _cross_entropy(a, y2)
    squared_distance2 = _squared_distance(a, y2)
    
    min_entropy_index2 = np.argmin(cross_entropy2)
    min_entropy_value2 = cross_entropy2[min_entropy_index2]
    min_entropy_a2 = a[min_entropy_index2]
    
    min_distance_index2 = np.argmin(squared_distance2)
    min_distance_value2 = squared_distance2[min_distance_index2]
    min_distance_a2 = a[min_distance_index2]

    plt.figure(figsize=(8, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(a, cross_entropy1,label='Cross Entropy '.format(y1), color='blue')
    plt.plot(a, squared_distance1, label='Squared Distance'.format(y1), color='red')
    plt.scatter(min_entropy_a1, min_entropy_value1, color='blue')
    plt.scatter(min_distance_a1, min_distance_value1, color='red')
    plt.xlabel('a')
    plt.ylabel('Value')
    plt.title('y={:.1f}'.format(y1))
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(a, cross_entropy2,label='Cross Entropy '.format(y1),color='blue')
    plt.plot(a, squared_distance2, label='Squared Distance'.format(y1),color='red')
    plt.scatter(min_entropy_a2, min_entropy_value2, color='blue')
    plt.scatter(min_distance_a2, min_distance_value2, color='red')
    plt.xlabel('a')
    plt.ylabel('Value')
    plt.title('y={:.1f}'.format(y2))
    plt.legend()

    plt.tight_layout()
    plt.show()

_plot_crs(a, y1=0.5, y2=0.8)
```
